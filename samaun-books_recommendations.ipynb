```python
# We are trying to create a recommendation system for our online store www.samaun.com for the people who buy books from us.
# To provide them recommendations that they may be intrested in. we have used our own data for this analysis.
import pandas as pd

df = pd.read_csv('orders_export_Samaun_books.csv')


#Convert now table's column into rows to get the users name who bought these books and setup customer-id as index.

df = df.set_index('Customer_ID') 


df = df.pivot(columns= 'Product_name',values='Product_price')

df.fillna(0, inplace=True)
# Tidy up the data and bring the table in 0/1 format for further analysis
df = df.replace(75.0,1)

df = df.replace(110.0,1)
df = df.replace(115.0,1)
df = df.replace(120.0,1)
df = df.replace(125.0,1)
df = df.replace(165.0,1)
df = df.replace(175.0,1)
df = df.replace(185.0,1)
df = df.replace(195.0,1)
df = df.replace(200.0,1)

df = df.replace(229.0,1)
df = df.replace(245.0,1)
df = df.replace(265.0,1)
df = df.replace(270.0,1)
df = df.replace(335.0,1)
df = df.replace(425.0,1)
df = df.replace(440.0,1)
df = df.replace(500.0,1)

#print(df)

#df.head()
# tidy up column names
df.rename(columns={'Khuchkandi- book by Garh Ratna Narendra singh negi' : 'Khuchkandi', 'Nainital- A historical and descriptive account by Jim Clay' : 'Nanital', 'Nanda ke Jagar by Basanti Bisht' : 'NandaJagar', 'Pauranik Uttarakhand (History)  by Dr Rakesh Mohan Nautiyal' : 'PauranikHistory', 'State of Uttrakhand development and governance' : 'stateofUttarakhand','Uttarakhand : Samagra Gyankosh by Dr Rajendra Prasad Balodi (IVth Edition)-Hindi' : 'SamagarGyanlosh', 'Uttarakhand Handbook by Shailendra Singh Negi' : 'UKHandbook', 'Uttarakhand Himalaya ki Rangpa Janjati By Alam singh Chauhan' : 'RangpaJanjati', 'Uttarakhand Ka Naveen Itihas by Dr Yashwant Singh Katoch-Winsar Publishing' : 'NaveenIthiaas', 'Uttarakhand Yearbook 2018 (Hindi) by Winsar Publishing' : 'Yearbook2018', 'Winsar Uttarakhand Atlas by Dr. Indu Bharti' : 'UttarakhandAtlas'}, inplace=True)
import numpy as np
# create a function to calculate cosine similarity between two vectors
def getCosine(x,y) :
    cosine = np.sum(x*y) / (np.sqrt(np.sum(x*x)) * np.sqrt(np.sum(y*y)))
    return cosine

#cerate a similarity matrix placeholder listing item vs. item

similarity_matrix = pd.DataFrame(index=df.columns, columns = df.columns)
similarity_matrix = similarity_matrix.fillna(np.nan)

similarity_matrix.ix[0:10,0:10]
    
# loop through the each column and calculate the cosine similarity between all items
for i in similarity_matrix.columns:
    for j in similarity_matrix.columns:
        #fill the placeholder with cosine similarities
        similarity_matrix.ix[i,j]= getCosine(df[i].values,df[j].values)
        
similarity_matrix.ix[0:15,0:15]


#Create a nearest neighbours placeholder for each item
data_nearest_neighbours = pd.DataFrame (index=df.columns,columns=['top_'+str(i+1) for i in range(5)])
data_nearest_neighbours = data_nearest_neighbours.fillna(np.nan)
data_nearest_neighbours.head()


# get the top 5 neighbours 5 each item
for i in data_nearest_neighbours.index:
    top_ten = pd.DataFrame(similarity_matrix.ix[i,]).sort_values([i],ascending=False) [1:6].index.values
    data_nearest_neighbours.ix[i,:] = top_ten
    
data_nearest_neighbours.head()

```
